```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------------------------
# 1.    (     )
# -         (#) .
# --------------------------------------------------------------------

# 2.     
# --------------------------------------------------------------------
#    2  .
def smooth_curve_function(x):
    #     .
    a = -20/21
    b = -1/21
    c = 1
    return a*x**2 + b*x + c

#     x  .
# -0.2 1.2 200     .
x_values = np.linspace(-0.2, 1.2, 200)

#  x  y   .
y_values = smooth_curve_function(x_values)

#      .
given_points_x = [0, 1, 0.7]
given_points_y = [1, 0, 0.5]


# --------------------------------------------------------------------
# 3.  
# --------------------------------------------------------------------
#   .
plt.figure(figsize=(10, 6))

#   . (label    )
plt.plot(x_values, y_values, label=r' $f(x) = -\frac{20}{21}x^2 - \frac{1}{21}x + 1$')

#       .
# zorder=5     .
plt.scatter(given_points_x, given_points_y, color='red', s=100, zorder=5, label='  ')

#     .
plt.title('   2  ', fontsize=16)
plt.xlabel('x ', fontsize=12)
plt.ylabel('y ', fontsize=12)

#  .
plt.legend(fontsize=12)

#  (grid)   .
plt.grid(True, linestyle='--', alpha=0.6)

# x=0, y=0     .
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

#    .
plt.show()
```

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

# --------------------------------------------------------------------
# 1.     
# --------------------------------------------------------------------
def equations(vars):
    a, k, d = vars
    #  (k>0)        
    if k <= 0:
        return [1e9, 1e9, 1e9]
        
    eq1 = a * np.log(k) + d - 1      # f(0) = 1
    eq2 = a * np.log(1 + k) + d - 0  # f(1) = 0
    eq3 = a * np.log(0.7 + k) + d - 0.5 # f(0.7) = 0.5
    return [eq1, eq2, eq3]

#      
initial_guess = [-1, 0.2, 1] 
a, k, d = fsolve(equations, initial_guess)

print("     :")
print(f"a = {a:.6f}")
print(f"k = {k:.6f}")
print(f"d = {d:.6f}")

# --------------------------------------------------------------------
# 2.    
# --------------------------------------------------------------------
#     
def log_function(x):
    return 0.59 * np.log(-x + 1.22) + 0.88

#    (     )
try:
    plt.rc('font', family='Malgun Gothic') # Windows
except:
    try:
        plt.rc('font', family='AppleGothic') # Mac
    except:
        plt.rc('font', family='sans-serif') # Linux/Other
plt.rcParams['axes.unicode_minus'] = False #    

#   x  ( x=-k     )
x_curve = np.linspace(-k + 0.001, 1.5, 400)
y_curve = log_function(x_curve)

points_x = [0, 1, 0.7]
points_y = [1, 0, 0.5]

#  
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, label=f' $f(x) = {a:.3f} \cdot \ln(x + {k:.3f}) + {d:.3f}$')
plt.scatter(points_x, points_y, color='red', s=100, zorder=5, label='  ')

#  
plt.axvline(x=-k, color='green', linestyle='--', label=f' x = {-k:.3f}')

plt.title('     ', fontsize=16)
plt.xlabel('x ')
plt.ylabel('y ')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
# y       
plt.ylim(-1, 1.5) 
plt.show()
```

```python
def log_function1(x):
    return 0.59 * np.log(-x + 1.22) + 0.88

def log_function2(x):
    return -0.59 * np.log(-x + 1.22) + 0.12


#   x  ( x=-k     )
x_curve = np.linspace(0, 1, 100)
y_curve_1 = log_function1(x_curve)
y_curve_2 = log_function2(x_curve)


#  
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve_1)
plt.plot(x_curve, y_curve_2)



plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
