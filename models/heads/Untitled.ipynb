```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------------------------
# 1. 한글 폰트 설정 (그래프에 한글이 깨지지 않게 하기 위함)
# - 사용하시는 운영체제에 맞는 코드를 하나만 남기고 나머지는 지우거나 주석처리(#) 하세요.
# --------------------------------------------------------------------

# 2. 함수 정의 및 데이터 생성
# --------------------------------------------------------------------
# 세 점을 지나는 2차 함수를 정의합니다.
def smooth_curve_function(x):
    # 이 값들은 이전에 계산된 계수입니다.
    a = -20/21
    b = -1/21
    c = 1
    return a*x**2 + b*x + c

# 부드러운 곡선을 그리기 위해 x값의 범위를 생성합니다.
# -0.2부터 1.2까지 200개의 점을 만들어 곡선을 매끄럽게 표현합니다.
x_values = np.linspace(-0.2, 1.2, 200)

# 생성된 x값들에 대한 y값을 함수를 통해 계산합니다.
y_values = smooth_curve_function(x_values)

# 문제에서 주어진 세 점을 리스트로 만듭니다.
given_points_x = [0, 1, 0.7]
given_points_y = [1, 0, 0.5]


# --------------------------------------------------------------------
# 3. 그래프 그리기
# --------------------------------------------------------------------
# 그래프의 크기를 설정합니다.
plt.figure(figsize=(10, 6))

# 함수 곡선을 그립니다. (label에 수식을 넣어 범례에 표시)
plt.plot(x_values, y_values, label=r'함수 $f(x) = -\frac{20}{21}x^2 - \frac{1}{21}x + 1$')

# 주어진 세 점을 빨간색 점으로 그래프에 표시합니다.
# zorder=5는 점이 선보다 위에 보이게 합니다.
plt.scatter(given_points_x, given_points_y, color='red', s=100, zorder=5, label='주어진 세 점')

# 그래프의 제목과 축의 이름을 설정합니다.
plt.title('세 점을 지나는 2차 함수 곡선', fontsize=16)
plt.xlabel('x 축', fontsize=12)
plt.ylabel('y 축', fontsize=12)

# 범례를 표시합니다.
plt.legend(fontsize=12)

# 배경에 격자무늬(grid)를 추가하여 가독성을 높입니다.
plt.grid(True, linestyle='--', alpha=0.6)

# x=0, y=0 축을 더 잘 보이게 강조합니다.
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

# 완성된 그래프를 화면에 보여줍니다.
plt.show()
```

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

# --------------------------------------------------------------------
# 1. 연립방정식 정의 및 계수 찾기
# --------------------------------------------------------------------
def equations(vars):
    a, k, d = vars
    # 진수 조건(k>0) 위배 시 큰 오류 값을 반환하여 계산을 방지
    if k <= 0:
        return [1e9, 1e9, 1e9]
        
    eq1 = a * np.log(k) + d - 1      # f(0) = 1
    eq2 = a * np.log(1 + k) + d - 0  # f(1) = 0
    eq3 = a * np.log(0.7 + k) + d - 0.5 # f(0.7) = 0.5
    return [eq1, eq2, eq3]

# 안정적인 해를 찾기 위한 초기 추정값
initial_guess = [-1, 0.2, 1] 
a, k, d = fsolve(equations, initial_guess)

print("수치해석을 통해 찾은 계수는 다음과 같습니다:")
print(f"a = {a:.6f}")
print(f"k = {k:.6f}")
print(f"d = {d:.6f}")

# --------------------------------------------------------------------
# 2. 함수 정의 및 시각화
# --------------------------------------------------------------------
# 찾은 계수로 최종 함수를 정의
def log_function(x):
    return 0.59 * np.log(-x + 1.22) + 0.88

# 한글 폰트 설정 (그래프에 한글이 깨지지 않게 하기 위함)
try:
    plt.rc('font', family='Malgun Gothic') # Windows
except:
    try:
        plt.rc('font', family='AppleGothic') # Mac
    except:
        plt.rc('font', family='sans-serif') # Linux/Other
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 그래프를 그릴 x값의 범위 (점근선 x=-k 보다 약간 큰 값에서 시작)
x_curve = np.linspace(-k + 0.001, 1.5, 400)
y_curve = log_function(x_curve)

points_x = [0, 1, 0.7]
points_y = [1, 0, 0.5]

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, label=f'함수 $f(x) = {a:.3f} \cdot \ln(x + {k:.3f}) + {d:.3f}$')
plt.scatter(points_x, points_y, color='red', s=100, zorder=5, label='주어진 세 점')

# 점근선 표시
plt.axvline(x=-k, color='green', linestyle='--', label=f'점근선 x = {-k:.3f}')

plt.title('세 점을 지나는 로그 함수 곡선', fontsize=16)
plt.xlabel('x 축')
plt.ylabel('y 축')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
# y축 범위를 조절하여 그래프가 더 잘 보이게 함
plt.ylim(-1, 1.5) 
plt.show()
```

```python
def log_function1(x):
    return 0.59 * np.log(-x + 1.22) + 0.88

def log_function2(x):
    return -0.59 * np.log(-x + 1.22) + 0.12


# 그래프를 그릴 x값의 범위 (점근선 x=-k 보다 약간 큰 값에서 시작)
x_curve = np.linspace(0, 1, 100)
y_curve_1 = log_function1(x_curve)
y_curve_2 = log_function2(x_curve)


# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve_1)
plt.plot(x_curve, y_curve_2)



plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
